```python
s = "Hello world!"
str(s)  # str() is a built-in function that converts the object to a string
repr(s)  # repr() returns any object, print() actually uses str()
str(1/7)
x = 10 * 3.25
y = 200 * 200
s = 'The value of x is ' + repr(x) + ', and y is ' + repr(y)
print(s)
hello = "Hello world"
h = 'The value of hello is ' + repr(hello)
print(h)
hellos = repr(hello)
print(hellos)
repr((x, y, ("spam", "eggs")))
```

```python
# 格式化字符串字面值
import math
print(f"The value of pi is approximately {math.pi:.10f}")
```

```python
table = {"Sjoerd": 4127, "Jack": 4098, "Dcab": 8637678}
for name, phone in table.items():
    print(f"{name:10} ==> {phone:10d}")
    # {name:10}: 10表示字段最小宽度为 10 个字符, 未显式指定对齐方式时，字符串默认左对齐，不足部分用空格在右侧补齐
    # {phone:10d}: d表示按十进制整数格式化, 10同样表示字段最小宽度为 10 个字符，整数默认右对齐，不足部分用空格在左侧补齐
```

```python
animals = "eels"
print(f"My favorite hovercraft is full of {animals}.")
print(f"My favorite hovercraft is full of {animals!r}.")

a = "hello\nworld"
print(f"{a!s}")  # str()
print(f"{a!r}")  # repr()
print(f"{a!a}")  # ascii()
```

```python
bugs = 'roaches'
count = 13
area = 'living room'
print(f'Debugging {bugs=} {count=} {area=}')
```

```python
# str.format()方法
print('We are the {} who say "{}!"'.format('knights', 'Ni'))
print('{0} and {1}'.format('spam', 'eggs'))
print('{1} and {0}'.format('spam', 'eggs'))
print('This {food} is {adjective}.'.format(
      food='spam', adjective='absolutely horrible'))
print('The story of {0}, {1}, and {other}.'.format('Bill', 'Manfred',other='Georg'))
table = {'Sjoerd': 4127, 'Jack': 4098, 'Dcab': 8637678}
print('Jack: {0[Jack]:d}; Sjoerd: {0[Sjoerd]:d}; '
      'Dcab: {0[Dcab]:d}'.format(table))
for x in range(1, 11):
    print('{0:2d} {1:3d} {2:4d}'.format(x, x*x, x*x*x))
```
